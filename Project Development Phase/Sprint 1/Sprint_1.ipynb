```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   zoom_range=0.2, 
                                   horizontal_flip=True)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
xtrain = train_datagen.flow_from_directory('Food/Train data/',
                                           target_size=(64,64),
                                           class_mode='categorical',
                                           batch_size=100)
```

```python
xtest = test_datagen.flow_from_directory('Food/Testdata',
                                         target_size=(64,64),
                                         class_mode='categorical',
                                         batch_size=100)
```
